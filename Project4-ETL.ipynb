```python
#  Import and read the charity_data.csv.
import pandas as pd
Loan_df = pd.read_csv("LoanApprovalPrediction.csv")
Loan_df.head()
```

```python
#Drop Nan values
Loan_df = Loan_df.dropna()
```

```python
#Replace Gender 
mapping = {'Male':1, 'Female':0}
Loan_df['Gender'] = Loan_df['Gender'].map(mapping)
```

```python
#Replace Loan Status 
mapping = {'Y':1, 'N':0}
Loan_df['Loan_Status'] = Loan_df['Loan_Status'].map(mapping)
```

```python
#Replace Married
mapping = {'Yes':1, 'No':0}
Loan_df['Married'] = Loan_df['Married'].map(mapping)
Loan_df['Self_Employed'] = Loan_df['Self_Employed'].map(mapping)
```

```python
#Replace Education 
mapping = {'Graduate':1, 'Not Graduate':0}
Loan_df['Education'] = Loan_df['Education'].map(mapping)
```

```python
#Replace Property Area 
mapping = {'Urban':2, 'Semiurban':1, 'Rural':0}
Loan_df['Property_Area'] = Loan_df['Property_Area'].map(mapping)
```

```python
# Dropping Loan_ID column
Loan_df.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
output_file_path = 'ETL_Completed_LoanApprovalPrediction.csv'
# Export the DataFrame to a CSV file
Loan_df.to_csv(output_file_path, index=False)
# Printing a message indicating successful export
print(f"DataFrame 'Loan_df' successfully exported to '{output_file_path}' as a CSV file.")
```
