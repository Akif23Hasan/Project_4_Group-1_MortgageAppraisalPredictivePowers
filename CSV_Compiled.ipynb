```python
#Import Dependencies
import pandas as pd
import os
```

```python
# Accommodate raw path to variables
raw_customer, raw_orders = "ShoppingCartData/customers.csv", "ShoppingCartData/orders.csv"
raw_products, raw_sales = "ShoppingCartData/products.csv", "ShoppingCartData/sales.csv"

# Read-in data
customer, order = pd.read_csv(raw_customer), pd.read_csv(raw_orders)
product, sales_data = pd.read_csv(raw_products), pd.read_csv(raw_sales)
```

```python
cust_order = pd.merge(left=customer, right=order, 
                      left_index=True, right_index=True) # merging
cop_data = pd.merge(left=cust_order, right=product, 
                    left_index=True, right_index=True) # merging
```

```python
#Create CSV of the complete data
cop_data.to_csv('cop_data.csv', index=False)

print("Dataframe 'cop_data' has been saved to 'cop_data.csv'")
```
